```python
# created a 2d plot for better personal understanding

def logistic_cost_2d(X,y):
    """ plots logistic cost """
    wx, by = np.meshgrid(np.linspace(-6,12,50),
                         np.linspace(0, -20, 40))
    points = np.c_[wx.ravel(), by.ravel()]
    cost = np.zeros(points.shape[0],dtype=np.longdouble)

    for i in range(points.shape[0]):
        w,b = points[i]
        cost[i] = compute_cost_matrix(X.reshape(-1,1), y, w, b, logistic=True, safe=True)
    cost = cost.reshape(wx.shape)

    fig = plt.figure(figsize=(9,5))
    fig.canvas.toolbar_visible = False
    fig.canvas.header_visible = False
    fig.canvas.footer_visible = False 

    ax = fig.subplots(2,2)
    fig.suptitle("W and X Costs")
    ax[0,0].plot(wx,cost)
    ax[0,0].plot(wx[0],cost[0])
    ax[0,0].plot(wx[-1], cost[-1])
    ax[0,0].set_ylabel("COST")

    ax[1,0].plot(wx,np.log(cost))
    ax[1,0].plot(wx[0],np.log(cost[0]))
    ax[1,0].plot(wx[-1],np.log(cost[-1]))
    ax[1,0].set_ylabel("log(COST)")
    ax[1,0].set_xlabel("WX")
    

    ax[0,1].plot(by,cost)

    ax[1,1].plot(by,np.log(cost))
    ax[1,1].set_xlabel("Bias Y")
    

    plt.show()
    return cost
```

```python
cst = logistic_cost_2d(x_train,y_train)
```
